a. Vector Operations

a.1 Scalar Multiplication of a Vector

```python
def scalar_multiplication(scale, vector):
  scaled_vector = []
  vector_dimension = len(vector)
  print("Vector Dimension: ", vector_dimension)
  for i in range(0, vector_dimension):
    current_vector_component = vector[i]
    print("Current Vector Component: ", current_vector_component)
    # scale each component of the current vector
    scaled_current_vector_component = scale * current_vector_component
    print("Scaled Current Vector Component: ", scaled_current_vector_component)
    scaled_vector.append(scaled_current_vector_component)
  return scaled_vector

input_vector = [3, 4, 2]
scale = 2
scaled_vector = scalar_multiplication(scale, input_vector)
print("Example Result 1: ", scaled_vector)

input_vector = [7, 9, 6]
scale = 15
scaled_vector = scalar_multiplication(scale, input_vector)
print("Example Result 2: ", scaled_vector)
```

a.2.1 Addition of two vectors

```python
def add_two_vectors(a, b):
  len_a = len(a)
  len_b = len(b)
  result_vector = []
  if (len(a) != len(b)):
    print("Vectors are not the same length")
    return
  for i in range(0, len_a):
    # A vector can be broken down into components along x, y, z axis in case of 3D space.
    # Add the vectors along the broken down components. (For example, Add the x component of A and the x component of vector B. Doing the same in case of y and z components)
    current_added_component = a[i] + b[i]
    result_vector.append(current_added_component)
  return result_vector

added_two_vectors = add_two_vectors([1, 3, 1], [2, 2, 2])

print("Example Result 1: ", added_two_vectors)
added_two_vectors = add_two_vectors([100, 7, 3], [110, 99, 300])
print("Example Result 2: ", added_two_vectors)

# negative test case
added_two_vectors = add_two_vectors([4, 5], [3])
```

a.2.2 Subtraction of two vectors

```python
def subtract_two_vectors(a, b):
  len_a = len(a)
  len_b = len(b)
  result_vector = []
  if (len(a) != len(b)):
    print("Vectors are not the same length")
    return
  for i in range(0, len_a):
    # Subtract the vectors along the broken down components. (For example, Subtract the x component of A and the x component of vector B. Doing the same in case of y and z components)
    current_added_component = a[i] - b[i]
    result_vector.append(current_added_component)
  return result_vector

subtracted_two_vectors = subtract_two_vectors([1, 3, 1], [2, 2, 2])
print("Example Result 1: ", subtracted_two_vectors)

subtracted_two_vectors = subtract_two_vectors([100, 7, 3], [110, 99, 300])
print("Example Result 2: ", subtracted_two_vectors)

# negative test case
subtracted_two_vectors = subtract_two_vectors([4, 5], [3])
```

2.3 Dot Product of two vectors of length n

```python
def compute_dot_product(a, b):
  if (len(a) != len(b)):
    print("Vectors are not the same length")
    return
  sum = 0
  for i in range(0, len(a)):
    # Multiply the each component of the two vectors.
    # For example, multiply the x component of vector A with the x component of vector B. Doing the same in case of the y and z components.

    # As per the formula, once all the components are multiplied, they have to be added together. Here, I am adding the multiplied components to the variable sum.
    sum = sum + a[i] * b[i]
  return sum



computed_dot_product = compute_dot_product([3, 5, 2], [-1, 3, 0])
print("Example Result 1: ", computed_dot_product)

computed_dot_product = compute_dot_product([1, 16, -11], [-7, 15, -2])
print("Example Result 2: ", computed_dot_product)

# negative test case
compute_dot_product([4, 5, 6], [1])
```

2.4 Cross Product for 3D Vectors

```python
def determinant_2x2_matrix(matrix):
  # principle diagonal - non principle diagonal
  return matrix[0][0] * matrix[1][1] - matrix[1][0] * matrix[0][1]

def compute_cross_product(a, b):
  if (len(a) != len(b)):
    print("Vectors are not the same length")
    return

  d1 = determinant_2x2_matrix([
      [a[1], a[2]],
      [b[1], b[2]]
  ])

  d2 = determinant_2x2_matrix([
      [a[0], a[2]],
      [b[0], b[2]]
  ])

  d3 = determinant_2x2_matrix([
      [a[0], a[1]],
      [b[0], b[1]]
  ])

  return [d1, -d2, d3]

computed_cross_product = compute_cross_product([3, 4, 5], [7, 8, 9])
print("Example Result 1: ", computed_cross_product)

computed_cross_product = compute_cross_product([2, 3, 4], [5, 6, 7])
print("Example Result 2: ", computed_cross_product)

compute_cross_product([2, -1], [7, 7, 7, 7])
```

b.2.1 Matrix Addition

```python
def matrix_addition(a, b):
    row_len_matrix_a = len(a)
    col_len_matrix_a = len(a[0])
    row_len_matrix_b = len(b)
    col_len_matrix_b = len(b[0])

    # check if the matrices have the same dimensions
    if row_len_matrix_a != row_len_matrix_b:
        print("Row lengths of both matrices are not the same.")
        return
    if col_len_matrix_a != col_len_matrix_b:
        print("Column lengths of both matrices are not the same.")
        return

    # initialize result_matrix with None values
    result_matrix = [[None for _ in range(col_len_matrix_a)] for _ in range(row_len_matrix_a)]

    # perform matrix addition
    for i in range(row_len_matrix_a):
        for j in range(col_len_matrix_a):
            result_matrix[i][j] = a[i][j] + b[i][j]

    return result_matrix

matrix_1 = [
  [1, 1, 1],
  [2, 1, 2],
  [1, 1, 1]
]
matrix_2 = [
  [1, 1, 1],
  [1, 1, 1],
  [1, 1, 1]
]
result = matrix_addition(matrix_1, matrix_2)
print("Example Result 1: ", result)

matrix_1 = [
  [6, 100, -9],
  [132, -77, 3],
  [8, -5, 22]
]
matrix_2 = [
  [3, -55, 1],
  [55, 6, -31],
  [2, -71, 88]
]
result = matrix_addition(matrix_1, matrix_2)
print("Example Result 2: ", result)

matrix_1 = [
  [1, 2, 1],
  [7, -1, 15]
]

matrix_2 = [
  [1, 3],
  [2, 7]
]

# Negative test case for column lengths
matrix_addition(matrix_1, matrix_2)

matrix_1 = [
    [1, 3, -15]
]

matrix_2 = [
    [7, 7, 6],
    [1, 9, 1]
]

# Negative test case for row lengths
matrix_addition(matrix_1, matrix_2)
```

b.2.1 Matrix Subtraction

```python
def matrix_subtraction(a, b):
    row_len_matrix_a = len(a)
    col_len_matrix_a = len(a[0])
    row_len_matrix_b = len(b)
    col_len_matrix_b = len(b[0])

    # check if the matrices have the same dimensions
    if row_len_matrix_a != row_len_matrix_b:
        print("Row lengths of both matrices are not the same.")
        return
    if col_len_matrix_a != col_len_matrix_b:
        print("Column lengths of both matrices are not the same.")
        return

    # initialize result_matrix with None values
    result_matrix = [[None for _ in range(col_len_matrix_a)] for _ in range(row_len_matrix_a)]

    # perform matrix addition
    for i in range(row_len_matrix_a):
        for j in range(col_len_matrix_a):
            result_matrix[i][j] = a[i][j] - b[i][j]

    return result_matrix

matrix_1 = [
  [1, 1, 1],
  [2, 1, 2],
  [1, 1, 1]
]
matrix_2 = [
  [1, 1, 1],
  [1, 1, 1],
  [1, 1, 1]
]
result = matrix_subtraction(matrix_1, matrix_2)
print("Example Result 1: ", result)

matrix_1 = [
  [6, 100, -9],
  [132, -77, 3],
  [8, -5, 22]
]
matrix_2 = [
  [3, -55, 1],
  [55, 6, -31],
  [2, -71, 88]
]
result = matrix_subtraction(matrix_1, matrix_2)
print("Example Result 2: ", result)

matrix_1 = [
  [1, 2, 1],
  [7, -1, 15]
]

matrix_2 = [
  [1, 3],
  [2, 7]
]

# Negative test case for column lengths
matrix_subtraction(matrix_1, matrix_2)

matrix_1 = [
    [1, 3, -15]
]

matrix_2 = [
    [7, 7, 6],
    [1, 9, 1]
]

# Negative test case for row lengths
matrix_subtraction(matrix_1, matrix_2)
```

b.3 Multiply the matrix with a vector of suitable size

```python
def transform_vector(matrix, vector):

  row_len_matrix = len(matrix)
  col_len_matrix = len(matrix[0])

  row_len_vector = len(vector)

  if col_len_matrix != row_len_vector:
    print("Number of columns in the matrix must be equal to the length of the vector.")
    return

  # initialize the result vector
  result_vector = [0 for _ in range(row_len_matrix)]

  for i in range(0, row_len_matrix):
    sum = 0
    for j in range(0, col_len_matrix):
        sum += matrix[i][j] * vector[j]
    result_vector[i] = sum

  return result_vector

matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]
vector = [1, 2, 3]

result = transform_vector(matrix, vector)
print("Example Result 1: ", result)

matrix = [
  [-4, 12],
  [7, -9]
]
vector = [7, 9]

result = transform_vector(matrix, vector)
print("Example Result 2: ", result)

matrix = [
  [7, -9, 13],
  [3, 6, 43],
  [-5, 2, -11]
]
vector = [8, 19]

# negative test case that tests the validity of matrix and vector lengths
transform_vector(matrix, vector)
```

b.4 Multiply the matrix with a suitable Matrix

```python
def matrix_multiplication(a, b):
  row_len_a = len(a)
  col_len_a = len(a[0])
  row_len_b = len(b)
  col_len_b = len(b[0])

  if col_len_a != row_len_b:
        print("Number of columns in the first matrix must be equal to the number of rows in the second matrix.")
        return

  # initialize the result matrix
  result = [[0 for _ in range(col_len_b)] for _ in range(row_len_a)]

  for i in range(0, row_len_a):
    for j in range(0, col_len_b):
      sum = 0
      # computing the element (of the final matrix) for each (i, j) values
      # k helps us to get the candidate elements that need to be considered to get the value of the element present at (i, j) of the final matrix
      for k in range(0, col_len_a):
        sum += a[i][k] * b[k][j]
      result[i][j] = sum

  return result

matrixA = [
  [1, 2, 3],
  [4, 5, 6],
  [2, 3, 9]
]

matrixB = [
  [2, 3, 5],
  [3, 4, 8],
  [1, 1, 1]]

result = matrix_multiplication(matrixA, matrixB)
print("Example Result 1: ", result)

matrixA = [
    [1, 2, 3],
    [4, 5, 6]
]
matrixB = [
    [7, 8],
    [9, 10],
    [11, 12]
]

result = matrix_multiplication(matrixA, matrixB)
print("Example Result 2: ", result)

# negative test case
matrixA = [[1, 2, 3]]
matrixB = [[4, 5, 6]]
matrix_multiplication(matrixA, matrixB)
```

```python
def determinant_2x2_matrix(matrix):
  # principle diagonal - non principle diagonal
  return matrix[0][0] * matrix[1][1] - matrix[1][0] * matrix[0][1]
determinant_2x2_matrix([[7, 3], [2, 9]])

def compute_adjoint_matrix(matrix):
  row_len_matrix = len(matrix)
  col_len_matrix = len(matrix[0])
  # initialize the result_matrix
  result_matrix = [[None for _ in range(col_len_matrix)] for _ in range(row_len_matrix)]
  # formula for the adjoint of a matrix [[a, b], [c, d]] is [[a, -b], [-c, a]]. The next four lines is responsible for building the adjoint matrix
  result_matrix[1][1] = matrix[0][0]
  result_matrix[0][0] = matrix[1][1]
  result_matrix[0][1] = - matrix[0][1]
  result_matrix[1][0] = - matrix[1][0]

  return result_matrix

def multiply_2x2_matrix_with_a_scalar(scale, matrix):
  row_len_matrix = len(matrix)
  col_len_matrix = len(matrix[0])
  # Multiply each element present in (i,j) with a scale
  for i in range(0, row_len_matrix):
    for j in range(0, col_len_matrix):
      matrix[i][j] = scale * matrix[i][j]
  return matrix

def inverse_2x2_matrix(matrix):
  adjoint_matrix = compute_adjoint_matrix(matrix)
  determinant = determinant_2x2_matrix(matrix)

  # test for singularity. If a matrix is not singular only then it's inverse exists.
  if determinant == 0:
        print("The matrix is singular and does not have an inverse.")
        return

  determinant_reciprocal = 1/determinant

  # Only for a 2x2 matrix, to get the invers we multiply the reciprocal of the determinant with the adjoint matrix
  result_matrix = multiply_2x2_matrix_with_a_scalar(determinant_reciprocal, adjoint_matrix)
  return result_matrix

result = inverse_2x2_matrix([[3, 2], [1, 5]])
print("Example 1: ", result)

result = inverse_2x2_matrix([[1, 1], [1, 1]])
```
